```python
#1.Carga tu archivo de Titanic.csv en un dataframe.
#importar la liberia
import pandas as pd
#llamar a el archivo csv
df = pd.read_csv('titanic.csv')
#llamar a la variable df
df
```

```python
#2..Muestra los nombres de las columnas y tipo de datos.

#obtiene la información
df.info()
```

```python
#3.Muestra las edades únicas registradas.

#se usa para observar todos los valores únicos
df['Age'].unique()
```

```python
#4.¿Cuántos pasajeros iban en el barco?

#se utiliza para contabilizar
df['Name'].count()
```

```python
#5.¿Cuál es la edad máxima registrada?

#se usa para encontrar el valor máximo
df['Age'].max()
```

```python
#6.¿Cuál es la edad mínima registrada?

#se usa para el valor minimo
df['Age'].min()
```

```python
#7.Obtener toda la información de los pasajeros ordenados de forma descendente por edad.

#ordena los valores
df.sort_values('Age',ascending=False)
```

```python
#8.¿Cuántas personas sobrevivieron?

#usa para sumar las personas sobrevivientes
df['Survived'].sum()
```

```python
#9.¿Cuántas personas murieron?

#se usa para calcular las personas que murieron
df['Name'].count() - df['Survived'].sum()
```

```python
#10.Muestra las estadísticas descriptivas de la columna edad.

#se utiliza para obtener estadísticas descriptivas
df['Age'].describe()
```
